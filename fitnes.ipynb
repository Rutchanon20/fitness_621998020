```python
import numpy as np
from numpy import random


class Teachers:
    def __init__(self):
        self.Tname = {
            "Pumipong",  # 0
            "Atikhom",  # 1
                        "Kittikorn",  # 2
                        "Mayoon",  # 3
                        "Thanawut"  # 4
        }

    def getTeacherName(self):
        return self.Tname


class Rooms:
    def __init__(self):
        self.Rname = {
            "4911",  # 0
            "4912",  # 1
            "4914",  # 2
            "4921",  # 3
            "4922"  # 4
        }
        self.SlotRoom = np.array([
            ["xxxxxxx"], ["xxxxxxx"],
            ["xxxxxxx"], ["xxxxxxx"],
            ["xxxxxxx"], ["xxxxxxx"],
            ["xxxxxxx"], ["xxxxxxx"],
            ["xxxxxxx"], ["xxxxxxx"]
        ])
        self.SlotRoomWeek = np.array([self.SlotRoom for _ in range(len(self.Rname))]  # Ex . for five rooms
                                     )

    def getRoomsName(self):
        return self.Rname

    def getRoomWeek(self):
        return self.SlotRoomWeek


class CoursesWorkload:
    def __init__(self):
        self.CWL = np.array([   # Course , Teacher , Room, TimeSlot
            ["CPE1111", 0, -1, -1],
            ["CPE1112", 0, -1, -1],
            ["CPE1113", 1, -1, -1],
            ["CPE2111", 1, -1, -1],
            ["CPE2112", 2, -1, -1],
            ["CPE2113", 2, -1, -1],
            ["CPE2114", 3, -1, -1],
            ["CPE2115", 3, -1, -1],
            ["CPE2100", 4, -1, -1],
            ["CPE2101", 4, -1, -1],
            ["CPE2102", 0, -1, -1],
            ["CPE2103", 1, -1, -1]
        ])

    def getCourseWorkload(self):
        return self.CWL

```

```python
# ------ Main Class ------


def Scheduling(Cwl, Rm):
    for i in range(len(Cwl)):
        rdRoom = random.randint(0, len(Rm)-1)
        rdSlot = random.randint(0, len(Rm[0])-1)
        while (Rm[rdRoom, rdSlot] != "xxxxxxx"):
            rdRoom = random.randint(0, len(Rm)-1)
            rdSlot = random.randint(0, len(Rm[0])-1)
        Cwl[i, 2] = rdRoom  # Cwl[i,2] = random.randint(0,len(Rm)-1)
        Cwl[i, 3] = rdSlot  # Cwl[i,3] = random.randint(0,len(Rm[0])-1)
        # Rm[int(Cwl[i,2]), int(Cwl[i,3])] = Cwl[i,0]
        Rm[int(Cwl[i, 2]), int(Cwl[i, 3])] = Cwl[i, 0]


def ComputeFittnessValue(Rm, Cwl):
    value = 50

    for i in range(len(Cwl)):
        for j in range(len(Cwl)):
            if i != j:
                if(Cwl[i, 3] == Cwl[j, 3]):
                    if(Cwl[i, 1] == Cwl[j, 1]):
                        return 0

    for r in range(len(Rm)):
        if (Rm[r, 6] != "xxxxxxx"):
            value -= 1
        if (Rm[r, 7] != "xxxxxxx"):
            value -= 1

    return value


def DisplayRoom(Rm):
    for i in range(len(Rm)):
        print()
        for sl in range(len(Rm[i])):
            print(Rm[i, sl], end=" ")
            if sl % 2 == 1:
                print()


def DisplayCourseWorkLoad(cwl):
    print(cwl)


value = 10000
for i in range(value):
    # np.random.seed(0)
    T = Teachers()
    R = Rooms()
    C = CoursesWorkload()
    Scheduling(C.getCourseWorkload(), R.getRoomWeek())
    value = ComputeFittnessValue(R.getRoomWeek(), C.getCourseWorkload())
    if value == 50:
        DisplayRoom(R.getRoomWeek())
        print(value)
        DisplayCourseWorkLoad(C.getCourseWorkload())
```
